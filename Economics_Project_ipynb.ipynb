# **Rank 3 out of (sedgwick,buttler,cowley) counties**

*   Butler County, KS â†’ 20015

```python
# List of area_fips codes for the counties and state
area_fips_codes = ['20000', '20173', '20015', '20035']

# Filter the DataFrame for the specified counties and state
filtered_df = df[df['area_fips'].astype(str).isin(area_fips_codes)]

# Display the filtered DataFrame
filtered_df[['area_fips','avg_annual_pay']]
```

```python
# Create a dictionary to map area_fips codes to county names
county_names = {

    '20173': 'Sedgwick County',
    '20015': 'Butler County',
    '20035': 'Cowley County'
}

# Create a new column with county names
filtered_df['county_name'] = filtered_df['area_fips'].astype(str).map(county_names)

# Sort the DataFrame by average annual pay in descending order
ranked_df = filtered_df.sort_values('avg_annual_pay', ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
barplot = sns.barplot(x='county_name', y='avg_annual_pay', data=ranked_df, palette="viridis")

# Add annotations on top of each bar
for index, row in ranked_df.iterrows():
    barplot.annotate(
        f"${row['avg_annual_pay']:,.0f}",  # Format as currency
        xy=(row['county_name'], row['avg_annual_pay']),
        xytext=(0, 5),  # Offset the text slightly above the bar
        textcoords='offset points',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

# Customize plot
plt.title('Ranking of the Kansas Counties-sedgwick,buttler,cowley by Average Annual Pay in Manufacturing')
plt.xlabel('County')
plt.ylabel('Average Annual Pay')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a dictionary to map area_fips codes to county names
county_names = {
    '20000': 'Kansas (Statewide)',
    '20173': 'Sedgwick County',
    '20015': 'Butler County',
    '20035': 'Cowley County'
}

# Create a new column with county names
filtered_df['county_name'] = filtered_df['area_fips'].astype(str).map(county_names)

# Separate Kansas (Statewide) for comparison
statewide_df = filtered_df[filtered_df['county_name'] == 'Kansas (Statewide)']
county_df = filtered_df[filtered_df['county_name'] != 'Kansas (Statewide)']

# Sort the DataFrame by average annual pay in descending order (only counties)
ranked_df = county_df.sort_values('avg_annual_pay', ascending=False)

# Create the bar plot for counties
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
barplot = sns.barplot(x='county_name', y='avg_annual_pay', data=ranked_df, palette="viridis")

# Add annotations on top of each county bar
for index, row in ranked_df.iterrows():
    barplot.annotate(
        f"${row['avg_annual_pay']:,.0f}",  # Format as currency
        xy=(row['county_name'], row['avg_annual_pay']),
        xytext=(0, 5),  # Offset the text slightly above the bar
        textcoords='offset points',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

# Add Kansas (Statewide) average as a horizontal line for comparison
if not statewide_df.empty:
    statewide_pay = statewide_df['avg_annual_pay'].values[0]
    plt.axhline(y=statewide_pay, color='red', linestyle='--', label=f'Kansas Statewide: ${statewide_pay:,.0f}')
    plt.legend()

# Customize plot
plt.title('Comparison of Kansas Counties (Sedgwick, Butler, Cowley) with Statewide Average Annual Pay in Manufacturing')
plt.xlabel('County')
plt.ylabel('Average Annual Pay')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

# **Kansas-manufacturing-private_ownership-Employment-2014 t0 2023**

```python
df_2014  = pd.read_csv('/content/kansas_2014.csv')
df_2015  = pd.read_csv('/content/kansas_2015.csv')
df_2016  = pd.read_csv('/content/kansas_2016.csv')
df_2017  = pd.read_csv('/content/kansas_2017.csv')
df_2018  = pd.read_csv('/content/kansas_2018.csv')
df_2019  = pd.read_csv('/content/kansas_2019.csv')
df_2020  = pd.read_csv('/content/kansas_2020.csv')
df_2021  = pd.read_csv('/content/kansas_2021.csv')
df_2022  = pd.read_csv('/content/kansas_2022.csv')
df_2023  = pd.read_csv('/content/kansas_2023.csv')
```

```python
df_2014.drop_duplicates(inplace=True)
df_2015.drop_duplicates(inplace=True)
df_2016.drop_duplicates(inplace=True)
df_2017.drop_duplicates(inplace=True)
df_2018.drop_duplicates(inplace=True)
df_2019.drop_duplicates(inplace=True)
df_2020.drop_duplicates(inplace=True)
df_2021.drop_duplicates(inplace=True)
df_2022.drop_duplicates(inplace=True)
df_2023.drop_duplicates(inplace=True)
```

```python
df_2023.dtypes
```

```python
df_2023.head()
```

```python
# Convert industry_code to numeric and handle non-numeric values
dfs = [df_2014, df_2015, df_2016, df_2017, df_2018,
       df_2019, df_2020, df_2021, df_2022, df_2023]

# Apply filtering with type conversion
for i, df in enumerate(dfs, start=2014):
    df['industry_code'] = pd.to_numeric(df['industry_code'], errors='coerce')  # Convert to numeric
    df['own_code'] = pd.to_numeric(df['own_code'], errors='coerce')           # Ensure own_code is numeric

# filter each DataFrame
df_2014_filtered = df_2014[(df_2014['own_code'] == 5) & (df_2014['industry_code'] == 1013)][['year', 'area_title', 'annual_avg_emplvl']]
df_2015_filtered = df_2015[(df_2015['own_code'] == 5) & (df_2015['industry_code'] == 1013)][['year', 'area_title', 'annual_avg_emplvl']]
df_2016_filtered = df_2016[(df_2016['own_code'] == 5) & (df_2016['industry_code'] == 1013)][['year', 'area_title', 'annual_avg_emplvl']]
df_2017_filtered = df_2017[(df_2017['own_code'] == 5) & (df_2017['industry_code'] == 1013)][['year', 'area_title', 'annual_avg_emplvl']]
df_2018_filtered = df_2018[(df_2018['own_code'] == 5) & (df_2018['industry_code'] == 1013)][['year', 'area_title', 'annual_avg_emplvl']]
df_2019_filtered = df_2019[(df_2019['own_code'] == 5) & (df_2019['industry_code'] == 1013)][['year', 'area_title', 'annual_avg_emplvl']]
df_2020_filtered = df_2020[(df_2020['own_code'] == 5) & (df_2020['industry_code'] == 1013)][['year', 'area_title', 'annual_avg_emplvl']]
df_2021_filtered = df_2021[(df_2021['own_code'] == 5) & (df_2021['industry_code'] == 1013)][['year', 'area_title', 'annual_avg_emplvl']]
df_2022_filtered = df_2022[(df_2022['own_code'] == 5) & (df_2022['industry_code'] == 1013)][['year', 'area_title', 'annual_avg_emplvl']]
df_2023_filtered = df_2023[(df_2023['own_code'] == 5) & (df_2023['industry_code'] == 1013)][['year', 'area_title', 'annual_avg_emplvl']]
```

```python
#  print all filtered data frames

print(df_2014_filtered)
print(df_2015_filtered)
print(df_2016_filtered)
print(df_2017_filtered)
print(df_2018_filtered)
print(df_2019_filtered)
print(df_2020_filtered)
print(df_2021_filtered)
print(df_2022_filtered)
print(df_2023_filtered)
```

```python

# Concatenate the filtered DataFrames
frames = [df_2014_filtered, df_2015_filtered, df_2016_filtered, df_2017_filtered,
          df_2018_filtered, df_2019_filtered, df_2020_filtered, df_2021_filtered,
          df_2022_filtered, df_2023_filtered]

Annual_emp_avg = pd.concat(frames)

# Display the result
Annual_emp_avg
```

```python

plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='annual_avg_emplvl', data=Annual_emp_avg, marker='o')

# Annotate each data point with its value
for x, y in zip(Annual_emp_avg['year'], Annual_emp_avg['annual_avg_emplvl']):
    plt.annotate(f'{y:.0f}',  # Format the value as needed
                 (x, y),
                 textcoords="offset points",
                 xytext=(0, 10),
                 ha='center')


plt.title('Trend of Annual Average Employment Level in Manufacturing (2014-2023)')
plt.xlabel('Year')
plt.ylabel('Annual Average Employment Level')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set Seaborn style for aesthetics
sns.set_style("whitegrid")

# Create the figure
plt.figure(figsize=(12, 6))

# Line plot with markers
sns.lineplot(
    x='year',
    y='annual_avg_emplvl',
    data=Annual_emp_avg,
    marker='o',
    color='royalblue',
    linewidth=2.5
)

# Annotate each data point with its value
for x, y in zip(Annual_emp_avg['year'], Annual_emp_avg['annual_avg_emplvl']):
    plt.annotate(f'{y:.0f}',
                 (x, y),
                 textcoords="offset points",
                 xytext=(0, 10),
                 fontsize=12,
                 ha='center')

# Add title and labels
plt.title('Trend of Annual Average Employment Level in Manufacturing (2014-2023)', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Annual Average Employment Level', fontsize=12)

# Ensure all years are displayed on the x-axis
plt.xticks(np.arange(2014, 2024, 1))

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Apply tight layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```
